```python
from IPython.display import HTML

# Defining the HTML and JavaScript for the animation
html_code = """
<div style='background: #222; padding: 10px; border-radius: 8px;'>
    <p style='color: white; font-family: sans-serif;'>Use <b>Arrow Keys</b> to move the circle!</p>
    <canvas id='animationCanvas' width='600' height='400' style='border:1px solid #555; background: #000;'></canvas>
</div>

<script>
(function() {
    const canvas = document.getElementById('animationCanvas');
    const ctx = canvas.getContext('2d');

    let x = canvas.width / 2;
    let y = canvas.height / 2;
    const radius = 20;
    const speed = 5;

    const keys = {};

    window.addEventListener('keydown', (e) => {
        keys[e.code] = true;
        // Prevent scrolling with arrows
        if(['ArrowUp','ArrowDown','ArrowLeft','ArrowRight'].includes(e.code)) e.preventDefault();
    });
    window.addEventListener('keyup', (e) => {
        keys[e.code] = false;
    });

    function update() {
        if (keys['ArrowUp'] && y > radius) y -= speed;
        if (keys['ArrowDown'] && y < canvas.height - radius) y += speed;
        if (keys['ArrowLeft'] && x > radius) x -= speed;
        if (keys['ArrowRight'] && x < canvas.width - radius) x += speed;
    }

    function draw() {
        ctx.clearRect(0, 0, canvas.width, canvas.height);

        // Draw Circle
        ctx.beginPath();
        ctx.arc(x, y, radius, 0, Math.PI * 2);
        ctx.fillStyle = '#00ffcc';
        ctx.shadowBlur = 15;
        ctx.shadowColor = '#00ffcc';
        ctx.fill();
        ctx.closePath();

        update();
        requestAnimationFrame(draw);
    }

    draw();
})();
</script>
"""

display(HTML(html_code))
```
